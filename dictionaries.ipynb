**1. Creating a dictionary**

```python
student = {
    "name": "Charlie",
    "age": 22,
    "course": "python"
}
print(student)
```

**2. Accessing Dictionary Values**

```python
print(student["name"])          #Access using keys
print(student.get("course"))       #get() is safer than []
```

**3. Adding & Updating Values**

```python
student["city"] = "Agra"   #ADD NEW KEY
print(student)

student["age"] = 23         #UPDATE EXISTING KEY
print(student)
```

**4. Removing Elements**

```python
student.pop("name")    # pop() removes key-value pair
print(student)

```

**5. Looping through a dictionary**

```python
for key in student:       #for keys
  print(key)

for value in student.values():       #for values
  print(value)

for key, value in student.items():     #for key-value pairs
  print(key, ":", value)
```

**6. Dictionary length**

```python
print("total keys:", len(student))
```

**7. Checking key existence**

```python
if "city" in student:            # Membership test
  print("city exists")
```

**8. Nested Dictionary**

```python
members = {
    "m1": {"name": "Ali", "dept": "IT"},
    "m2": {"name": "Bob", "dept": "HR"}
}
print(members["m2"]["dept"])
```
