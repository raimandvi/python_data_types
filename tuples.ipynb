**1. Creating a set**

```python
numbers = {10, 20, 30, 40}
print(numbers)
```

**2. Sei with duplicate elements**

```python
values = {1, 2, 2, 3, 3, 4}
print(values)
```

**3. Empty set**

```python
emp_set = set()  # if we use {} then it creates dictionary not set
print(type(emp_set))
```

**4. Adding, removing elements**

```python
# Adding a element to a set:
fruits = {"apple", "banana"}
fruits.add("cherry")
print(fruits)

# Removing element from a set
fruits.remove("banana")
print(fruits)
fruits.discard("mango")
print(fruits)
```

**5. Set operations(Union, intersection, difference)**

```python
a = {2,4,5}
b = {3,5,2}

print(a|b)
print(a&b)
print(a-b)
```

**6. Membership test**

```python
students = {'Ali', 'Bob', 'charlie'}

print('Ali' in students)
print('Dev' not in students)
```

**7. Converting list to set**

```python
numbers = [1,2,2,4,5,6,4,5,6,8,9]
set_list = set(numbers)
print(set_list)
```
