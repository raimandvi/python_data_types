**1. Creating a tuple**

```python
numbers = (5, 10, 15, 20, 25)
print(numbers)
```

**2. Accessing Elements**

```python
fruits = ("Apple","Banana","Mango","Guava")
print(fruits[1])
print(fruits[-1])
```

**3. Looping through a tuple**

```python
for i in fruits:
  print(i)
```

**4. tuple Length**

```python
print("Length of the tuple is: ", len(fruits))
```

**5. Tuple packing and unpacking**

```python
person = ("Sundar", 55, "Data Engineer")
name, age, role = person

print(name)
print(age)
print(role)
```

**6. Converting list to tuple**

```python
s_no = [1,2,3,4,5]
no_tuple = tuple(s_no)
print(no_tuple)
```
