**1. Creating a tuple**

```python
numbers = (10, 20, 30, 40)
print(numbers)
```

**2. Accessing tuple elements**

```python
fruits = ("apple","banana","cherry")
print(fruits[0])
print(fruits[-1])
```

**3. Tuple Length**

```python
numbers = (5,10,20)
print("Length of tuple: ", len(numbers))
```

**4. Tuple Packing and Unpacking**

```python
student = ("Angel",22, "Data Engineer")

name, age, role = student

print(name)
print(age)
print(role)
```

**5. Converting list to tuple**

```python
numbers = [1,2,3,4,5]
num_tuple = tuple(numbers)
print(num_tuple)
```
