**1. Creating a List**

```python
numbers = [5, 10, 15, 20, 25]
print(numbers)
```

**2. Accessing, Adding, Removing Elements**

```python
print(numbers[1])
print(numbers[-1])

numbers.append(11)
numbers.insert(0, 4)
print(numbers)

numbers.remove(20)
numbers.pop()
print(numbers)
```

**3. Looping through a list**

```python
for i in numbers:
  print(i)
```

**4. List Length**

```python
print("Length of the list is: ", len(numbers))
```
